```python
import os

def set_project_rood_dir():
    notebook_dir = os.path.dirname(os.path.abspath("__file__"))
    paths = notebook_dir.split("/")

    # remove directory unless notebook/s directory is found
    while len(paths) > 0:
        if paths[-1] == 'notebook' or paths[-1] == 'notebooks':
            paths.pop()
            break
        paths.pop()

    # show error if paths is empty
    if len(paths) == 0:
        print("Current directory: ", notebook_dir)
        raise ValueError("Unable to find notebook/s directory in path")

    root = "/".join(paths)
    os.chdir(root)
    print("Successfully changed working directory: ", root)
    print("Current working directory: ", os.getcwd())


set_project_rood_dir()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA_PATH = "data/raw/train.csv"
TEST_DATA_PATH = "data/raw/test.csv"
```

```python
df = pd.read_csv(DATA_PATH, index_col=0)
test_df = pd.read_csv(TEST_DATA_PATH, index_col=0)
```

```python
# qq plot
import scipy.stats as stats
import statsmodels.api as sm

def plot_qqplot(data, title):
    sm.qqplot(data, line='s')
    plt.title(title)
    plt.show()
```

```python
for col in df.columns:
    plot_qqplot(df[col], col)
```

```python
# kde plot

for col in df.columns:
    print("\n\n")
    print(col)
    sns.kdeplot(df[col])
    print("skew:",df[col].skew())
    print("kurtosis:",df[col].kurtosis())

    plt.title(col)
    plt.show()
```

```python
# violin plot

for col in df.columns:
    print("\n\n")
    print(col)
    sns.violinplot(df[col])
    plt.title(col)
    plt.show()
```

```python
# box plot

for col in df.columns:
    print("\n\n")
    print(col)
    sns.boxplot(df[col])
    plt.title(col)
    plt.show()
```
